```python
%matplotlib inline
import quantstats as qs
import pandas as pd
import numpy as np
from scipy.stats import ttest_1samp
from pandas.tseries.offsets import BDay
```

```python
indice_list = ['DAX Index', 'OMX Index']
# Load price data for the pair
pair_name = indice_list[0] + " - " + indice_list[1]
returns_df = pd.read_csv(r"/content/drive/MyDrive/IAQF 2023/Reruns/new_results/DAX Index_OMX Index__portfolio.csv")
returns_df["Dates"] = pd.to_datetime(returns_df['Dates'], format = "%Y-%m-%d")
returns_df.set_index("Dates", inplace = True)
portfolio_value = returns_df*1000000 + 1000000
qs.reports.html(portfolio_value['portfolio_returns'], benchmark = portfolio_value['baseline_returns'], rf = 0.02, ouput = True, title = '{0}'.format(pair_name), download_filename='det_{0}.html'.format(pair_name), benchmark_title = 'S&P')
```
